```python
# tokenization :: preprocessing
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(oov_token = '<OOV>')
```

```python
tokenizer.fit_on_texts(data)
```

```python
# tokenizer.word_index
sequences = tokenizer.texts_to_sequences(data)
```

```python
sequences[0]
```

```python
input_sequence = []

for sequence in sequences:
  # n_GRAM
  for i in range(1, len(sequence)):
    n_gram = sequence[: i + 1]
    input_sequence.append(n_gram)
```

```python
input_sequence
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
max_len = max([len(x) for x in input_sequence])
```

```python
padded_input = pad_sequences(input_sequence, maxlen = max_len)
```

```python
x_train  = padded_input[:, :-1]
y_train = padded_input[:, -1]
```

```python
total_words = len(tokenizer.word_index) + 1
```

```python
total_words
```

```python
import tensorflow as tf
```

```python
# one hot encoding
y_train = tf.keras.utils.to_categorical(y = y_train, num_classes = total_words)
```

```python
y_train.shape
```

```python
y_train[0].shape
```

```python
# model implementation
# RNN - Recurrent Neural Network

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim = total_words, output_dim = 240, input_length = max_len - 1),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(150)),
    tf.keras.layers.Dense(units = total_words, activation = 'softmax')
])
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate = 0.008)  # try with lr = 0.01, 0.001, 0.0008, 0.0005, 0.0001
model.compile(loss = 'categorical_crossentropy', optimizer = optimizer, metrics = ['accuracy'])
```

```python
history = model.fit(x = x_train, y = y_train, epochs= 15, batch_size= 50)   # replace 30 with 15 epochs
```

```python
# Hello ARJVAUO hello, i am A I chat bot. How may i help you . OUAVJRA

def predict(question):

  next_word = ''
  while next_word != 'ouavjra':
    # preprocessing
    question_sequence = tokenizer.texts_to_sequences([question])
    padded_question = pad_sequences(question_sequence, maxlen = max_len)

    # prediction
    prediction = model.predict(x = padded_question)

    # index of most_probable word
    index_of_most_probable_word = np.argmax(prediction)
    # index to word coversion
    predicted_word = tokenizer.index_word[index_of_most_probable_word]

    question += predicted_word

    next_word = predicted_word


  answer = question.split('arjvauo')[1].split('ouavjra')[0]

  return answer
```

```python
question = 'Hello'
predict(question)
```
